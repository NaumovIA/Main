```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('populations.txt')
data
```

```python
dataT = data.T
```

```python
#1 Plot the populations for the years in the period. Add legend and axis labels to the plot.
plt.grid()
plt.plot(dataT[0], dataT[1], label = 'Hare')
plt.plot(dataT[0], dataT[2], label = 'Lynx')
plt.plot(dataT[0], dataT[3], label = 'Carrot')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(bbox_to_anchor=(1.2, 1), loc='upper left', borderaxespad=0.)
```

```python
#2 Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each species for the years in the period.
plt.grid()
plt.plot(dataT[0], dataT[1]/(dataT[1] + dataT[2]) * 100, label = 'Hare')
plt.plot(dataT[0], dataT[2]/(dataT[1] + dataT[2]) * 100, label = 'Lynx')
plt.xlabel('Year')
plt.ylabel('%')
plt.legend(bbox_to_anchor=(1.2, 1), loc='upper left', borderaxespad=0.)


```

```python
 #3 Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Arrange 3 subplots in a row.
hare_mean = np.mean(dataT[1])
lynx_mean = np.mean(dataT[2])
carrot_mean = np.mean(dataT[3])
hare_std = np.std(dataT[1])
lynx_std = np.std(dataT[2])
carrot_std = np.std(dataT[3])
print('hare mean = ', hare_mean)
print('lynx mean = ', lynx_mean)
print('carrot mean = ', carrot_mean)
print('hare std = ', hare_std)
print('lynx std = ', lynx_std)
print('carrot std = ', carrot_std)   
```
